```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import netCDF4 as nc
```

```python
region_numbers = {"Mid-Atlantic Bight": 0, 
                  "Southern New England": 1, 
                  "Georges Bank": 2, 
                  "Gulf of Maine": 3, 
                  "Bay of Fundy": 4, 
                  "Scotian Shelf": 5, 
                  "Off-Shelf": 6, 
                  np.nan: 7}
```

```python
def get_data(species):
    data_base = pd.read_csv("../data/model_sims/baseline/{}/settled_data.csv".format(species))
    data_base = data_base.drop(columns=["Unnamed: 0"])

    data_warm = pd.read_csv("../data/model_sims/warm/{}/settled_data.csv".format(species))
    data_warm = data_warm.drop(columns=["Unnamed: 0"])
    
    return data_base, data_warm
```

```python
def get_settled(a, s):
    asett = np.ma.MaskedArray(a, mask=s)
    return asett[asett.mask].data

def get_settled_particles(x, y, s):
    xsett = get_settled(x, s)
    ysett = get_settled(y, s)
    
    return xsett, ysett
```

# Atlantic Cod

```python
data_base, data_warm = get_data("atlantic-cod")
```

```python
xsb, ysb, ssb = data_base['final_lons'].to_numpy(), data_base['final_lats'].to_numpy(), data_base['settled_particles'].to_numpy()
xsw, ysw, ssw = data_warm['final_lons'].to_numpy(), data_warm['final_lats'].to_numpy(), data_warm['settled_particles'].to_numpy()
```

```python
xsett_base, ysett_base = get_settled_particles(xsb, ysb, ssb)
hb, xe, ye = np.histogram2d(xsett_base, ysett_base, bins=300)

xsett_warm, ysett_warm = get_settled_particles(xsw, ysw, ssw)
hw, xe, ye = np.histogram2d(xsett_warm, ysett_warm, bins=300)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
im1 = ax1.imshow(np.log10(hb.T), interpolation='nearest', origin='lower', extent=[xe[0], xe[-1], ye[0], ye[-1]], cmap="jet", vmin=0, vmax=5)
ax1.set_xlabel("Longitude")
ax1.set_ylabel("Latitude")

im2 = ax2.imshow(np.log10(hw.T), interpolation='nearest', origin='lower', extent=[xe[0], xe[-1], ye[0], ye[-1]], cmap="jet", vmin=0, vmax=5)
ax2.set_xlabel("Longitude")
ax2.set_ylabel("Latitude")

cbaxes1 = fig.add_axes([0.49, 0.25, 0.01, 0.50])
cbaxes2 = fig.add_axes([0.91, 0.25, 0.01, 0.50])

cb = plt.colorbar(im1, ax=ax1, cax = cbaxes1)
cb = plt.colorbar(im2, ax=ax2, cax = cbaxes2)
```
