```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from collections import Counter
```

```python
def get_data(species):
    data_base = pd.read_csv("../data/model_sims/baseline/{}/settled_data.csv".format(species))
    data_base = data_base.drop(columns=["Unnamed: 0"])

    data_warm = pd.read_csv("../data/model_sims/warm/{}/settled_data.csv".format(species))
    data_warm = data_warm.drop(columns=["Unnamed: 0"])
    
    return data_base, data_warm
```

```python
def get_region_counts(data):
    init_region_counts = Counter(data[data['settled_particles'] == 1.0]['init_region'])
    final_region_counts = Counter(data[data['settled_particles'] == 1.0]['final_region'])
    return init_region_counts, final_region_counts
```

```python
def get_count_diff(inits, finals):

    diff_count = {}
    for r in finals.keys():
        init = inits[r]
        final = finals[r]
        diff_count[r] = (final - init)/init
    return diff_count
```

```python
def plot_population_diffs(base, warm, title):    
    labels = ['Mid-Atlantic Bight', 'Southern New England', 'Georges Bank', 'Gulf of Maine', 'Bay of Fundy', 'Scotian Shelf']
    base_vals = []
    warm_vals = []
    for l in labels:
        base_vals.append(base[l]*100)
        warm_vals.append(warm[l]*100)

    x = np.arange(len(labels))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    rects1 = ax.bar(x - width/2, base_vals, width, label='Baseline')
    rects2 = ax.bar(x + width/2, warm_vals, width, label='Warm')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_title(title)
    ax.set_ylabel('Percent Change')
    ax.set_xticks(x)
    #ax.set_ylim(-1.2, 1.2)
    ax.set_xticklabels(labels, rotation=45)
    ax.legend()
    
    return fig, ax
```

# Atlantic Cod

```python
data_base, data_warm = get_data("atlantic-cod")
```

```python
base_init_regions, base_final_regions = get_region_counts(data_base) 
warm_init_regions, warm_final_regions = get_region_counts(data_warm) 
```

```python
base_diff = get_count_diff(base_init_regions, base_final_regions)
warm_diff = get_count_diff(warm_init_regions, warm_final_regions)
```

```python
fig, ax = plot_population_diffs(base_diff, warm_diff, title="Proportional Population Change of Atlantic Cod")
```
