```python
import pandas as pd
import numpy as np
```

```python
fname = "../auxdata/spawning-predictions.csv"
```

```python
data = pd.read_csv(fname)
```

```python
data.drop(["Unnamed: 0"], axis=1, inplace=True)
data
```

```python
cod_data = data[data['Species'] == "Atlantic Cod"]
cod_data = cod_data[cod_data['month'] < 6]
cod_data.describe()
```

```python
cod_data['predicted'].sum()
```

```python
species_ym_sums = {}
for fish in ['Atlantic Cod', 'Haddock', "Yellowtail Flounder", 'Atlantic Mackerel', 'American Butterfish']:
    print(fish)
    year_mo_sums = {}
    for y in range(1980, 1990):
        print(y)
        for m in range(1, 12):
            mo = str(m).zfill(2)
            key = str(y)+mo
            df = data[data['Species'] == fish]
            year_mo_sums[key] = df[(df['year'] == y) & (df['month'] == m)]['predicted'].sum()
    species_ym_sums[fish] = year_mo_sums
    
species_ym_sums
```

```python
for fish in ['Atlantic Cod', 'Haddock', "Yellowtail Flounder", 'Atlantic Mackerel', 'American Butterfish']:
    for y in range(1980, 1990):
        mo = "12"
        key = str(y)+mo
        df = data[data['Species'] == fish]
        species_ym_sums[fish][key] = df[(df['year'] == y) & (df['month'] == 12)]['predicted'].sum()
```

```python
species_ym_sums['Haddock']
```

```python
def std_pred(row):
    y = row['year']
    m = row['month']
    f = row['Species']
    s = species_ym_sums[f][str(y)+str(m).zfill(2)]
    return row['predicted']/s
```

```python
cod_data['pred_std'] = cod_data.apply(std_pred, axis=1)
```
